```python
import requests
import json
import pandas as pd
import urllib.parse
```

```python
API_Key = 
Destinations_list=open('dests.txt')
Target=[]
Longitude=[]
Latitude=[]
Distance_km=[]
Duration=[]
parms={}
for line in Destinations_list:
    address=line
    Target.append(line.rstrip())
    url="https://maps.googleapis.com/maps/api/geocode/json?address=%s&key=%s" % (address,API_Key)
    try:
        response = requests.get(url).json()
        if not response['results'][0]["status"] == "OK":
            print("HTTP error")
        else:
            try:
                response = json.loads(response)
            except:
                print("Response not in valid JSON format")
    except:
        Longitude.append(response['results'][0]["geometry"]["location"]['lng'])
        Latitude.append(response['results'][0]["geometry"]["location"]['lat'])  
        service_url="https://maps.googleapis.com/maps/api/distancematrix/json?origins=tel%20aviv&"
        parms['destinations']=line
        parms['units']='kilometers'
        parms['key']=API_Key
        url=service_url+urllib.parse.urlencode(parms)
    try:
        response = requests.get(url).json()
        if not response['results'][0]["status"] == "OK":
            print("HTTP error")
        else:
            try:
                response = json.loads(response)
            except:
                print("Response not in valid JSON format")
    except:
        Distance_km.append(response['rows'][0]["elements"][0]["distance"]["text"])
        Duration.append(response['rows'][0]["elements"][0]["duration"]["text"])
        
df = pd.DataFrame(list(zip(Target,Distance_km,Duration,Longitude,Latitude)),columns =["Target", "Distance_km", "Duration","Longitude","Latitude"])
```
